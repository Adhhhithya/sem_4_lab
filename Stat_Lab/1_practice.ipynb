```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
true_mean = 5.0
true_std = 2.0
```

```python
np.random.seed(42)
n_samples=1000
data=np.random.normal(true_mean,true_std,n_samples)
```

```python
estimated_mean =np.mean(data)
```

```python
fisher_info=n_samples/(true_std**2)
crlb=1/fisher_info
```

```python
print("True mean(U):",true_mean)
print("Estimated mean(U^):",estimated_mean)
print("Fisher Information(I(U)):",fisher_info)
print("Cramer-Rao Lower Bound(CRLB):",crlb)
```

```python
plt.figure(figsize=(10,6))
plt.hist(data,bins=30,alpha=0.7,label="Generated Data")
plt.axvline(true_mean,color='red',linestyle='--',label=f"True Mean:{true_mean}")
plt.axvline(estimated_mean,color='green',linestyle='--',label=f"Estimated Mean:{estimated_mean:.2f}")
plt.title("Gaussian Distribution with Estimated Mean")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid()
plt.legend()
plt.show()
```
