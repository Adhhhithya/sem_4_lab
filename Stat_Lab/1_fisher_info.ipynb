```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define parameters for the Gaussian distribution
true_mean = 5.0  # Actual mean
true_std = 2.0   # Standard deviation (known)
```

```python
# Generate synthetic data
np.random.seed(42)  # For reproducibility
n_samples = 1000  # Number of samples
data = np.random.normal(true_mean, true_std, n_samples)
```

```python
# Estimate the mean from the data
estimated_mean = np.mean(data)
```

```python
# Fisher Information for the mean of a Gaussian distribution
# I(μ) = n / σ^2, where n = number of samples, σ = standard deviation
fisher_information = n_samples / (true_std**2)
```

```python

# Calculate variance of the estimated mean (Cramer-Rao Lower Bound)
crlb = 1 / fisher_information
```

```python
print("True Mean (μ):", true_mean)
print("Estimated Mean (μ̂):", estimated_mean)
print("Fisher Information (I(μ)):", fisher_information)
print("Cramer-Rao Lower Bound (Variance of μ̂):", crlb)
```

```python
# Visualization
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, alpha=0.7, label="Generated Data")
plt.axvline(true_mean, color='red', linestyle='--', label=f"True Mean (μ): {true_mean}")
plt.axvline(estimated_mean, color='green', linestyle='--', label=f"Estimated Mean (μ̂): {estimated_mean:.2f}")
plt.title("Gaussian Distribution with Parameter Estimation")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid()
plt.show()
```
